```python
def country_Raise_temp(Raise_temp, country):
    Raise_temp1 = Raise_temp[(Raise_temp['element'] == 'Temperature change') & (Raise_temp['area'] == country)]
    Raise_temp1 = Raise_temp1.set_index('months').transpose()[2:]
    Raise_temp1['year'] = Raise_temp1.index
    Raise_temp1.reset_index(drop=True, inplace=True)
    Raise_temp1.index.names = [country]
    Raise_temp1 = Raise_temp1.astype('float')
    return Raise_temp1
```

```python
def seasons_Raise_temp(Raise_temp, country):
    Raise_temp1 = Raise_temp[(Raise_temp['element'] == 'Temperature change') & (Raise_temp['area'] == country)]
    Raise_temp1.rename(columns={'months': 'seasons'}, inplace=True)
    Raise_temp1 = Raise_temp1.set_index('seasons').transpose()[2:]
    Raise_temp1['year'] = Raise_temp1.index
    Raise_temp1.drop(columns=months, inplace=True)
    Raise_temp1.reset_index(drop=True, inplace=True)
    Raise_temp1.index.names = [country]
    Raise_temp1= Raise_temp1.astype('float')
    return Raise_temp1
```

```python
def country_plot(data, period):
    p = plt.figure(figsize=(8,8))
    sns.regplot(data=data, x='year', y=period, fit_reg=True, lowess=True, scatter_kws={'alpha':0.2}, line_kws={'lw':2, 'alpha':0.75})
    plt.ylabel('∆ °C', rotation=0) 
    plt.title(data.index.name)
```

```python
country_plot(India,'January')
country_plot(China,'January')
country_plot(USA,'January')
```
